```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn import metrics
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_excel("S4_2.xls")
df
```

```python
df.info()
```

```python
X = df[["sbp", "waist", "ca", "phos", "job", "home", "j"]]
Y = df[["bmi"]]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
predictions = model.predict(X_test)
```

```python

plt.figure(figsize=(10, 6))
plt.scatter(Y_test, predictions, color='blue', label='Predicted Values')
plt.scatter(Y_test, Y_test, color='green', label='Actual Values', alpha=0.5)
plt.title('Actual vs Predicted BMI with Regression Line')
plt.legend()
plt.grid()
plt.show()
```
